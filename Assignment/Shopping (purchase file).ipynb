### Shopping

```python
import json
import time

fd = open("pro_records.json",'r')
sh = fd.read()
fd.close()
records = json.loads(sh)

print("-------------PRODUCT LIST:------------- \n ")
records
```

```python

print("***************************************************************")
print("----------------------Welcome----------------------------------")
print("***************************************************************")

u_prod = input("Enter the product Id to Buy          : ")
u_quan = int(input("Enter the quantity of the Product    :"))
dat      = time.ctime()

for i in records.keys():
    while u_prod == i:
        if u_quan <= records[i]['quan']:
            print("***************************************************************")
            print("-----------------------BILLING DETAILS:------------------------")
            print("***************************************************************")
            print(" product             :", records[u_prod]['name'])
            print(" price               :", records[u_prod]['price'])
            print(" Expiry date         :", records[u_prod]['expiry'])
            print(" Manufacturing Brand :", records[u_prod]['brand'])
            print(" Time                :", dat )
            print(" Total Amount        :", records[u_prod]['price']*u_quan)
            print("**************************************************************")
            print("*****************--Thank You. Visit Again--*******************")
            
            records[u_prod]["quan"] = records[u_prod]["quan"] - u_quan
        else:
            print("***************************************************************")
            print("Sorry for the inconvenience...Insufficient Stock Quantity")
            print("***************************************************************")
        break
if u_prod not in records.keys():
    print("***************************************************************")
    print(" Invalid Product ID     :",u_prod)
    print("  Enter Valid Product Id")
    print("***************************************************************")
```

### updating the pro_records.json file

```python
    
js = json.dumps(records)

fd = open("pro_records.json",'w')
fd.write(js)
fd.close()
```

### creating and updating sales.json file

```python

sales = {'product_id':u_prod,' product_name':records[u_prod]['name'], 'quantity' : u_quan, 'amount': records[u_prod]['price'] * u_quan, 'date/time':dat}
sale = json.dumps(sales)

fd = open("sales.json",'a')
fd.write(sale)
fd.close()

```
